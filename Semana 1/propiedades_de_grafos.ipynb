## Polinomios cromáticos
Definición: Sea $P$ el polinomio cromático de $G$. Entonces $P(k)$ representa el número de coloraciones posibles con con $k$ colores. 

Particularmente, el mínimo entero $k$ tal que $P(k)>0$ es el número cromático de $G$. 

Además, $|P(-1)|$ es la cantidad de orientaciones acíclicas de $G$. En donde una orientación es volver a cada arista del grafo una arista dirigida. 


```python
G = nx.read_adjlist('G_adylist.txt', nodetype=int)
nx.draw_networkx(G)
```

```python
p = nx.chromatic_polynomial(G)
for i in range(-1, len(G)):
    print(f'{i}: ' + str(p.subs({'x': i})))
```

```python
def chromatic_number(G):
    p = nx.chromatic_polynomial(G)
    for i in range(len(G)):
        if(p.subs({'x': i}) > 0):
            return i
    return -1
```
